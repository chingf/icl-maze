# Load Dataset and Create Environment

```python
n_eval_envs = 50

is_h5_file = eval_dset_path.endswith('.h5')
if is_h5_file:
    eval_trajs = h5py.File(eval_dset_path, 'r')
    traj_indices = list(eval_trajs.keys())
    n_eval_envs = min(n_eval_envs, len(traj_indices))
    random.seed(0)
    traj_indices = random.sample(traj_indices, n_eval_envs)
    random.seed()
    eval_trajs = [eval_trajs[i] for i in traj_indices]
else:  # Pickle file
    with open(eval_dset_path, 'rb') as f:
        eval_trajs = pickle.load(f)
    n_eval_envs = min(n_eval_envs, len(eval_trajs))
    random.seed(0)
    eval_trajs = random.sample(eval_trajs, n_eval_envs)
    random.seed()
```

```python
def run_model(traj, model, state_features, seq_length):
    hidden_states = [[] for _ in range(model.n_layer)]
    
    for state_feature in state_features: 
        batch = {
            'context_states': convert_to_tensor([np.array(traj['context_states'])]),
            'context_actions': convert_to_tensor([np.array(traj['context_actions'])]),
            'context_next_states': convert_to_tensor([np.array(traj['context_next_states'])]),
            'context_rewards': convert_to_tensor([np.array(traj['context_rewards'])[:, None]]),
            'query_states': convert_to_tensor([np.array(state_feature)]),
            }
        batch['zeros'] = torch.zeros(1, 10 ** 2 + 4 + 1).float()
        for k in batch.keys():
            if 'context' in k:
                batch[k] = batch[k][:,:seq_length]
            batch[k] = batch[k].to(model.device)
        model.save_activations = True
        with torch.no_grad():
            out = model(batch)
        _hidden_states = model.activations['hidden_states'][1:] # Tuple over layers
        for i_layer in range(model.n_layer):
            hidden_states[i_layer].append(_hidden_states[i_layer].cpu().numpy().squeeze()[-1])
    return hidden_states
```

```python
def get_node_and_representation(traj, env, model):
    nodes = []
    hidden_states = []
    seq_lengths = []
    dist_to_reward = []
    state_features = []
    state_features = list(env.node_map.keys())
    state_features = [s for s in state_features if list(s) != env.goal.tolist()]
    _, dist_from_goal = env.make_opt_action_dict()

    _nodes = [] 
    _dist_to_reward = []
    for state_feature in state_features:
        _nodes.append(str(env.node_map[state_feature]))
        _dist_to_reward.append(dist_from_goal[state_feature])
    for seq_length in [1, 25, 50, 75, 100, 250, 500, 750, 1000, 1100, 1200, 1300, 1400, 1500, 1600]:
        _hidden_states = run_model(traj, model, state_features, seq_length=seq_length)
        for l in range(model.n_layer):
            if len(_hidden_states[l]) != len(_nodes):
                print(l)
                print(len(_hidden_states[l]), len(_nodes))
                import pdb; pdb.set_trace()
        nodes.append(_nodes)
        hidden_states.append(_hidden_states)
        seq_lengths.append(seq_length)
        dist_to_reward.append(_dist_to_reward)

    return nodes, hidden_states, seq_lengths, dist_to_reward
```

```python
results = {
    'nodes': [],
    'hidden_states': [],
    'model_layer': [], 
    'context_length': [],
    'env': [], 
    'dist_to_reward': []
}

for i_eval, traj in enumerate(eval_trajs[:10]):
    env_config = {
        'max_layers': 7,
        'horizon': 1600,
        'branching_prob': 1.0,
        'node_encoding_corr': corr,
        'state_dim': state_dim,
        'initialization_seed': np.array(traj['initialization_seed']).item()
    }
    env = EvalCntrees().create_env(env_config, np.array(traj['goal']), i_eval)
    nodes, hidden_states, context_lengths, dist_to_reward = get_node_and_representation(traj, env, model)
    for _nodes, _hidden_states, _context_length, _dist_to_reward in zip(nodes, hidden_states, context_lengths, dist_to_reward):
        for i_layer, model_layer_hidden_states in enumerate(_hidden_states):
            results['nodes'].extend(_nodes)
            results['hidden_states'].extend(model_layer_hidden_states)
            results['model_layer'].extend([i_layer] * len(_nodes))
            results['context_length'].extend([_context_length] * len(_nodes))
            results['env'].extend([i_eval] * len(_nodes))
            results['dist_to_reward'].extend(_dist_to_reward)
```

```python
df = pd.DataFrame(results)
```

```python
same_dist_to_reward = {
    'sim': [],
    'model_layer': [],
    'context_length': [],
    'comparison_node_dist_to_reward': []
}
diff_dist_to_reward = {
    'sim': [],
    'model_layer': [],
    'context_length': [],
    'comparison_node_dist_to_reward': []
}
envs = np.sort(df['env'].unique())

for context_length in df['context_length'].unique():
    context_length_df = df[df['context_length'] == context_length]
    for i_layer in context_length_df['model_layer'].unique():
        layer_df = context_length_df[context_length_df['model_layer'] == i_layer]

        # Get goal-leaf similarity
        for env in envs:
            env_df = layer_df[layer_df['env'] == env]
            other_env_df = layer_df[layer_df['env'] != env]

            for dist_to_reward in np.sort(layer_df['dist_to_reward'].unique()):
                _env_df = env_df[env_df['dist_to_reward'] == dist_to_reward]
                _other_env_df_diff = other_env_df[other_env_df['dist_to_reward'] != dist_to_reward]
                _other_env_df_same = other_env_df[other_env_df['dist_to_reward'] == dist_to_reward]
                _env_reprs = np.stack(_env_df['hidden_states'].to_numpy())
                _other_env_diff_reprs = np.stack(_other_env_df_diff['hidden_states'].to_numpy())
                _other_env_same_reprs = np.stack(_other_env_df_same['hidden_states'].to_numpy())
                cc_diff = np.corrcoef(_env_reprs, _other_env_diff_reprs)
                cc_same = np.corrcoef(_env_reprs, _other_env_same_reprs)
                cc_diff = cc_diff[:_env_reprs.shape[0], _env_reprs.shape[0]:].flatten().tolist()
                cc_same = cc_same[:_env_reprs.shape[0], _env_reprs.shape[0]:].flatten().tolist()
                same_dist_to_reward['sim'].extend(cc_same)
                same_dist_to_reward['model_layer'].extend([i_layer] * len(cc_same))
                same_dist_to_reward['context_length'].extend([context_length] * len(cc_same))
                same_dist_to_reward['comparison_node_dist_to_reward'].extend([dist_to_reward] * len(cc_same))
                diff_dist_to_reward['sim'].extend(cc_diff)
                diff_dist_to_reward['model_layer'].extend([i_layer] * len(cc_diff))
                diff_dist_to_reward['context_length'].extend([context_length] * len(cc_diff))
                diff_dist_to_reward['comparison_node_dist_to_reward'].extend([dist_to_reward] * len(cc_diff))
```

```python
same_dist_to_reward_df = pd.DataFrame(same_dist_to_reward)
diff_dist_to_reward_df = pd.DataFrame(diff_dist_to_reward)
```

```python
_same_dist_to_reward_df = same_dist_to_reward_df[same_dist_to_reward_df['context_length'] == 1000]
_diff_dist_to_reward_df = diff_dist_to_reward_df[diff_dist_to_reward_df['context_length'] == 1000]
fig, ax = plt.subplots(figsize=(5, 3))
ax = sns.barplot(
    data=pd.concat([
        _same_dist_to_reward_df.assign(type='Same Reward Distance'),
        _diff_dist_to_reward_df.assign(type='Different Reward Distance')
    ]),
    x='model_layer', y='sim', hue='type', ax=ax, 
    )

for i, bar in enumerate(ax.patches):
    if bar.get_facecolor() == (0.7642156862745098, 0.5318627450980391, 0.12598039215686285, 1):  # C1 color
        bar.set_hatch('///')
    bar.set_facecolor('dimgray')
plt.ylabel('Pearson Correlation')
plt.xlabel('Model Layer')
plt.xticks(np.arange(model.n_layer), np.arange(model.n_layer)+1)
plt.legend()
plt.show()

```
